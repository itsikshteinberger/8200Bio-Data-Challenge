# Imbalanced data handling
As we have seen in the description the data is not balanced at all, as we deal with this problem.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Data/Clean_data_1.csv')
df2 = pd.read_csv('Data/Clean_data_2.csv')
```

```python
df1
```

```python
df2
```

We will use the easiest method as it works in many cases not bad at all:<br/>
OverSampling - Duplicate the data with the smaller amount in the required amount.
## OverSampling

```python
from tqdm import tqdm
 
for specialty in tqdm(df1.specialty.value_counts().keys()):
    num = int(979 / len(df1[df1['specialty'] == specialty].index)) - 1
    df = df1[df1['specialty'] == specialty] 
    df1 = pd.concat([df1] + [df]*num)
```

```python
for specialty in tqdm(df1.specialty.value_counts().keys()):
    num = int(979 / len(df2[df2['specialty'] == specialty].index)) - 1
    df = df2[df2['specialty'] == specialty] 
    df2 = pd.concat([df2] + [df]*num)
```

```python
print(df1.specialty.value_counts())
print('\n============\n')
print(df2.specialty.value_counts())
```

Pay attention! The data has grown significantly now,these are two large files - the process may take time.<br/>Patience is required :)

```python
df1.to_csv('Data/Clean_data_1.csv', index=False)
```

```python
df2.to_csv('Data/Clean_data_2.csv', index=False)
```
